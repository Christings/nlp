```python
import spacy
import pandas as pd
import itertools as it

nlp=spacy.load('en_core_web_md')
```

```python
%%time
with open('4095_01.txt',encoding='utf-8') as f:
    for doc in f:
#         print(doc)
        doc=nlp(doc)
        for num,entity in enumerate(doc.ents):
            print('Entity {}:'.format(num+1),entity,'-',entity.label_)
            print('')
        for token in doc:
            print(token.text,token.lemma_,token.pos_,token.tag_,token.dep_,token.shape_,token.is_alp)
```
